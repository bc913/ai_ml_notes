# Image Recognition

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Grayscale

```python
img_data = plt.imread('images/Abbey_Road.jpg')
```

```python
gray_img_data = np.average(img_data, weights=[0.3, 0.59, 0.11], axis=2)
```

```python
plt.figure(figsize=(10,10))
plt.subplot(1, 2, 1)
plt.imshow(img_data)
plt.subplot(1,2,2)
plt.imshow(gray_img_data, cmap='gray')
plt.show()
```
